```r
beta0 <- c(3, -1)
zi <- matrix(c(1:50/10, runif(50)), nrow=2, ncol=50, byrow=TRUE)
zi
```

```r
plot(zi[1,])
plot(zi[2,])
```

```r
plot(as.vector(crossprod(beta0, zi)))
```

```r
lambdai <- exp(as.vector(crossprod(beta0, zi)))
plot(lambdai)
```

```r
yi <- rexp(length(lambdai), rate=lambdai)
plot(yi)
```

```r
gyi <- sqrt(yi) # or other increasing function
plot(gyi)
```
